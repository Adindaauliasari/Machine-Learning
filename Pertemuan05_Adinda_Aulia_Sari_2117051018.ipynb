# Latihan 2

# Data set

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/student_score_pass.csv")
df2
```

```python
df2 = df2.dropna()
df2
```

# Preprocessing

```python
df_training =  df2.drop(['npm','pass_test'], axis=1)
y = [0 if x == "not pass" else 1 for x in df2["pass_test"]]

df_training
```

# Modeling

```python
clustering = AgglomerativeClustering().fit(df_training)
#hasil clustering
df_training["kluster"] = clustering.labels_
df_training
```

# Tugas

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_training)
model = AgglomerativeClustering(n_clusters=2, linkage="complete")
assigned_cluster = model.fit_predict(df_scaled)
print(assigned_cluster)
```

```python
df_training["cluster"] = assigned_cluster
df_training["true"] = y
df_training
```

```python
adjusted_rand_score(assigned_cluster, y)
```

```python
fowlkes_mallows_score(assigned_cluster, y)
```
